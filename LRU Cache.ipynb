```python
class Node:
    def __init__(self):
        self.key = None
        self.val = None
        self.previous = None
        self.next = None

class dqueue:
    
    def __init__(self, capacity:int):  
        self.capacity = capacity
        self.front = Node
        self.last = Node
        self.front.next = self.last
        self.last.previous = self.front

    def insert(self, k:int, v:int)->Node:
        node = Node()
        node.val = v
        node.key = k
        if(self.front.next == self.last):
            self.front.next = node
            self.last.previous = node
            node.previous = self.front
            node.next = self.last
        else:
            self.front.next.previous = node
            node.next = self.front.next
            self.front.next = node
            node.previous = self.front      
        return node

    def printQueue(self):
        print("Queue:")
        temp = self.front.next
        while(temp != self.last):
            print(temp.val)
            temp = temp.next

    def deleteNode(self, node):        
        node.next.previous = node.previous
        node.previous.next = node.next

    def get_last_node(self)->Node:
        if(self.last.previous == self.front):
            return None
        return self.last.previous 
    
    def put_at_front(self,node:Node):
        node.previous.next= node.next
        node.next.previous = node.previous
        self.front.next.previous = node
        node.next = self.front.next
        self.front.next = node
        node.previous = self.front
        

class LRUCache:
    
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.queue = dqueue(capacity)
        self.key_to_node_dict = dict()

    def get(self, key: int) -> int:
        if(key in self.key_to_node_dict):
            self.update_cache(key)
            return self.key_to_node_dict[key].val
        return -1
    
    def update_cache(self,key):
        node = self.key_to_node_dict[key]
        self.queue.put_at_front(node)
        return

    def put(self, key: int, value: int) -> None:
        if(key  in self.key_to_node_dict):
            self.key_to_node_dict[key].val = value
            self.update_cache(key)
            return
        if(len(self.key_to_node_dict) >= self.capacity):
            last_node = self.queue.get_last_node()
            print("Last Node:",last_node.val)
            if(last_node.key != None):              
            #self.key_to_node_dict[key].val = value
                del self.key_to_node_dict[last_node.key]
                self.queue.deleteNode(last_node)        
        
        first_node = self.queue.insert(key,value)
        self.key_to_node_dict[key] = first_node
        return
    def printDict(self):
        print()
        print(self.key_to_node_dict)
        print()
```
